```python
total_males=np.ma.array(np.concatenate([allmales['all']['Median income'][:].values,allmales['white']['Median income'][:].values,allmales['black']['Median income'][:].values,allmales['hispanic']['Median income'][:].values,allmales['asian']['Median income'][:].values]).flatten())

total_females=np.ma.array(np.concatenate([allfemales['all']['Median income'][:].values,allfemales['white']['Median income'][:].values,allfemales['black']['Median income'][:].values,allfemales['hispanic']['Median income'][:].values,allfemales['asian']['Median income'][:].values]).flatten())

#I am gonna create the mask int he next line. 
#the mask is True where either male or female value is Nan. 
#Notice that True is really just >1, and False really 0. 
#so i can add the male and female mask and the resulting mask will be 
#False where both arrays are not Nan, and true otherwise. 

allmask = [np.isnan(total_males) + np.isnan(total_females)]

total_females.mask = allmask
total_males.mask = allmask
```

```python
#fit and plot

all_males = []
all_females = []

for race in allfemales:
    all_females.append(allfemales[race]['Median income'][1])
    all_males.append(allmales[race]['Median income'][1])
all_females_array = np.array(all_females)
all_males_array = np.array(all_males)

#print arrayFemalePoint
#print arrayMalePoint

coefficients = np.polyfit(all_females_array,all_males_array,1)
polynomial = np.poly1d(coefficients)
newY = polynomial(all_females_array)

fig = pl.figure(figsize=(15,10))
ax1 = fig.add_subplot(111)
ax1.set_ylabel('Male Total Median Income')
ax1.set_xlabel('Female Total Median Income')
ax1.set_title('Median Incomes of Females vs Males by Type of Income')
ax1.set_ylim([0,50000])
ax1.set_xlim([0,50000])

ax1.plot(total_females, total_males, 'ro')


ax1.plot(all_females_array,newY, ":",linewidth=1,color='tomato',label='First regression line')

coefficients2 = [beta1,beta0]
polynomial2 = np.poly1d(coefficients2)
anotherY = polynomial2(all_females_array)

ax1.plot(all_females_array,anotherY,"--", color = 'burlywood', label='Second regression line')

ax1.legend()

print beta1
```
