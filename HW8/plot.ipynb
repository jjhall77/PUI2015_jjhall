```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%pylab inline
```

```python
data = pd.read_csv('emission.csv')
```

```python
fig, ax = plt.subplots()
colors = {'Metropolitan Statistical Area':'#7570b3', 
          'Non-MSA':'#d95f02', 
          'Micropolitan Statistical Area':'#1b9e77'}
#7570b3


grouped = data.groupby('metropolitan.micropolitan.statistical.area')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='popestimate2002', y='percap', label=key, 
               color=colors[key], alpha=0.6, s=3, figsize=(12,8))

ax.set_yscale('log')
ax.set_xscale('log')
ax.grid(b=False)
plt.ylim(0.00000001,10)
plt.tick_params(axis='both', which='both', 
                bottom='off', top='off', 
                labelbottom='on', 
                right='off', 
                left='off', 
                labelleft='on')
plt.ylabel('Carbon Emissions per Capita (million tons)')
plt.xlabel('County Population')
plt.title('Per Capita Carbon Emissions by U.S. County, 2002')



plt.show()
```
